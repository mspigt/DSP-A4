```python
import pandas as pd
import numpy as np
```

```python
file_23_cannabis = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=0)
file_23_cocaine = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=1)
file_23_amphetamines = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=2)
file_23_ecstasy = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=3)
file_23_prescription_stimulants = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=4)
file_23_opioids = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=5)
file_23_opiates = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=6)
file_23_prescription_opioids = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=7)
file_23_tranquillizers_and_sedatives = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=8)
file_23_NPS = pd.read_excel("20231.2_Prevalence_of_drug_use_in_the_general_population_including_NPS_national_data.xlsx",skiprows=2, sheet_name=9)
```

```python
file_23_cannabis
```

```python
def fill_in_regions(file_23, idx):
    last_region = ""
    for index in file_23.index:
        if not pd.isna(file_23.iloc[index, idx]):
            last_region = file_23.iloc[index, idx]
        else:
            file_23.iloc[index, idx] = last_region
    return file_23
```

```python
def clean_file(file, drug):
    filled_in_regions = fill_in_regions(file, 1)
    drop_non_europe = filled_in_regions[filled_in_regions["Region"]=="Europe"]
    drop_columns = drop_non_europe.iloc[:,3:-5]
    int_year = drop_columns.astype({"Year": int})
    drop_gender = int_year.drop(columns=["Male", "Female", "Low", "High"])
    year_after_2020 = drop_gender[drop_gender["Year"]>=2020]
    added_type = year_after_2020.assign(Metabolite=drug) 
    renamed = added_type.rename(columns={'Country/Territory': 'Country', 'Best':'Percentage of the population'})
    dropped_nan =  renamed.dropna()
    return dropped_nan.set_index(["Country", "Year", "Metabolite"])
```

```python
cannabis_df = clean_file(file_23_cannabis, "cannabis")
cocaine_df = clean_file(file_23_cocaine, "cocaine")
amphetamines_df = clean_file(file_23_amphetamines, "amphetamine")
ecstasy_df = clean_file(file_23_ecstasy, "MDMA")
prescription_stimulants_df = clean_file(file_23_prescription_stimulants, "Prescription Stimulants")
opioids_df = clean_file(file_23_opioids, "Opioids")
opiates_df = clean_file(file_23_opiates, "Opiates")
prescription_opioids_df = clean_file(file_23_prescription_opioids, "Prescription Opioids")
```

```python
def clean_file_t_and_s(file):
    filled_in_regions = fill_in_regions(file, 1)
    drop_non_europe = filled_in_regions[filled_in_regions["Region"]=="Europe"]
    drop_columns = drop_non_europe.iloc[:,3:-3]
    int_year = drop_columns.astype({"Year": int})
    drop_gender = int_year.drop(columns=["Male", "Female", "Age"])
    year_after_2020 = drop_gender[drop_gender["Year"]>=2020]
    added_type = year_after_2020.assign(Metabolite="Tranquillizers and sedatives") 
    renamed = added_type.rename(columns={'Country/Territory': 'Country', 'Total':'Percentage of the population'})
    dropped_nan =  renamed.dropna()
    return dropped_nan.set_index(["Country", "Year", "Metabolite"])
 
tranquillizers_and_sedatives_df = clean_file_t_and_s(file_23_tranquillizers_and_sedatives)
tranquillizers_and_sedatives_df
```

```python
def clean_file_nps(file):
    filled_in_regions = fill_in_regions(file, 2)
    drop_non_europe = filled_in_regions[filled_in_regions["Region"]=="Europe"]
    drop_columns = drop_non_europe.iloc[:,4:-2]
    drop_columns["Year"] = pd.to_numeric(drop_columns["Year"], errors='coerce')
    drop_columns = drop_columns.dropna(subset=["Year"])
    int_year = drop_columns.astype({"Year": int})
    drop_gender = int_year.drop(columns=["Unnamed: 7", "Unnamed: 8", "Past year", "Unnamed: 10",
                                         "Unnamed: 11", "Past Month", "Unnamed: 13", "Unnamed: 14"])
    year_after_2020 = drop_gender[drop_gender["Year"]>=2020]
    renamed = year_after_2020.rename(columns={'Country/Territory': 'Country', 'Lifetime':'Percentage of the population', 'Substance':'Metabolite'})
    renamed['Metabolite'] = renamed['Metabolite'].str.lower()
    dropped_nan =  renamed.dropna()
    return dropped_nan.set_index(["Country", "Year", "Metabolite"])
 
NPS_df = clean_file_nps(file_23_NPS)
NPS_df
```

```python
UN_df = pd.concat([cannabis_df, cocaine_df, amphetamines_df, ecstasy_df, prescription_stimulants_df, opioids_df, opiates_df, prescription_opioids_df, tranquillizers_and_sedatives_df, NPS_df], ignore_index=False,axis=0)
```

```python
UN_df.to_csv("../UN_world_drug_report.csv")
UN_df
```
