```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ww-data-long-2023-05-02.csv")
df.head()
```

```python
df2 = df.drop(columns=['Site ID', 'City', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday', 'Weekday mean', 'Weekend mean'])
df3 = df2[df2['Year']>= 2020]
```

```python
result_df = df3.groupby(["Country", "Year", "Metabolite"]).mean()
result_df
```

```python
#  (litre/day), per day per 1 000 inhabitants
result_df
```

```python
print("highest daily mean:")
print(result_df.idxmax())
print("lowest daily mean:")
print(result_df.idxmin())
print("overall mean")
print(result_df.mean())
```

```python
result_df = result_df.rename(columns={'Daily mean': 'litre/day per 1 000 inhabitants'})
result_df = result_df.assign(source="waste water") 
result_df
```

```python
result_df.to_csv("../waste_water.csv")
```
